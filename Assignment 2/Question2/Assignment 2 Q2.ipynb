```python
#Question2 part 2 split of time
row=[]
#print(data1)
row.append('Name,Country Code,Day of Week,Start Time Hour,End Time Hour,Start Time Minute, End Time Minute\n')
for each in res_files:
    with open(os.path.join(path_to_json1, each)) as json_file:
        data1 = json.load(json_file)
        if (data1['term']=='restaurants' and 'hours' in data1.keys()):
                var1=data1['name']
                var2=data1['location']['country']
                #print(data1['hours'][0]['open'][1])
                for i in range(len(data1['hours'][0]['open'])):
                    var3=data1['hours'][0]['open'][i]['day']
                    start_hour=data1['hours'][0]['open'][i]['start']
                    end_hour=data1['hours'][0]['open'][i]['end']
                    if(len(start_hour)==4):
                        start_hour_split_1=str(start_hour[:2])
                        start_hour_split_2=str(start_hour[2:])
                    elif(len(start_hour)==3):
                        start_hour_split_1=str(start_hour[:1])
                        start_hour_split_2=str(start_hour[2:])
                    if(len(end_hour)==4):
                        end_hour_split_1=str(end_hour[:2])
                        end_hour_split_2=str(end_hour[2:])
                    elif(len(end_hour)==3):
                        end_hour_split_1=str(end_hour[:1])
                        end_hour_split_2=str(end_hour[2:])
                    row.append(str(var1)+','+str(var2)+','+str(var3)+','+str(start_hour_split_1)+','+str(start_hour_split_2)+','+str(end_hour_split_1)+','+str(end_hour_split_2)+'\n')
                        
with open('restaurants_split.csv','w',encoding='utf-8',errors='ignore') as csv_file:
    for item in row:
        csv_file.write(item)                        
                        
                        
                    
            
        
```
